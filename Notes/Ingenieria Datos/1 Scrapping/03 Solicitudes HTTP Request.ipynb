# Request HTTP

Requiere importar request

```python
import requests
```

Se manda llamar un request get, el cual devuelve una respuesta. Podemos saber el tipo de objeto con "type"

```python
response = requests.get('http://www.platzi.com')
type(response) #Tipo de objeto
```

Podemos ver las funciones y atributos de un objeto con "dir"

```python
print(dir(response)) #Funciones implementadas
```

Entre los atributos está el valor de "status_code"

```python
print(response.status_code) #Estatus de solicitud
```

Otro atributo son los headers, el cual es un diccionario

```python
print(response.headers) #Headers de solicitud
```

```python
print(response.headers['Date']) #Header "Date" de solicitud
```

También podemos ver el archivo completo del response

```python
print(response.text) #Contenido de solicitud
```
